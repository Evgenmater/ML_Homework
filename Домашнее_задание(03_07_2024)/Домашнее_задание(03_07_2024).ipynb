### Задание 1: Анализ данных о студентах с использованием NumPy


Описание:
У вас есть данные о студентах, включающие оценки по различным предметам. Используйте библиотеку NumPy для анализа этих данных.

```python
students_data = {
    'math': [85, 78, 92, 70, 88],
    'physics': [90, 82, 76, 85, 89],
    'informatics': [88, 92, 80, 87, 90]
}

students_data
```

Задание:

Найдите среднюю оценку по каждому предмету.<br>
Найдите медианную оценку по каждому предмету.<br>
Вычислите стандартное отклонение по каждому предмету.<br>
Определите предмет с самой высокой средней оценкой.

```python
import numpy as np

arr_v = np.array(list(students_data.values()))
print('Cредняя оценка по каждому предмету', np.mean(arr_v, axis=1))
print('Медианная оценка по каждому предмету', np.median(arr_v, axis=1))
print('Стандартное отклонение по каждому предмету', np.var(arr_v, axis=1))
print('Предмет с самой высокой средней оценкой', np.mean(arr_v, axis=1).max())
```

### Задание 2: Анализ данных о продажах с использованием Pandas

Описание:
Используйте библиотеку Pandas для анализа данных о продажах товаров. У вас есть данные о продажах товаров за последний год.

```python
sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    'Sales': [15000, 17000, 16000, 18000, 14000, 19000, 22000, 21000, 23000, 25000, 24000, 26000]
}
sales_data
```

Задание: <br>
Найдите общий объем продаж за год.<br>
Найдите средний объем продаж в месяц.<br>
Найдите месяц с наибольшим и наименьшим объемом продаж.<br>
Постройте линейный график изменения объема продаж по месяцам.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(sales_data)

print('общий объем продаж за год: ', df['Sales'].sum())
print('средний объем продаж в месяц: ', df['Sales'].median().astype(int))
print('месяц с наибольшим продаж): ', df.loc[df['Sales'].idxmax(), 'Month'])
print('месяц с наименьшим объемом продаж: ', df.loc[df['Sales'].idxmin(), 'Month'])

print('\nЛинейный график изменения объема продаж по месяцам: ',)
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(df['Month'], df['Sales'], marker='o', color='blue', label='Объём продаж')
ax.set(
    title='Линейный график',
    xlabel='Месяц',
    ylabel='Объём продаж',
    ylim=(df['Sales'].min()//1.1, df['Sales'].max()*1.1)
)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Задание 3: Задание на визуализацию данных с использованием Matplotlib и Seaborn

Описание:
Используйте готовый набор данных из библиотеки Seaborn для создания различных типов графиков. Ваша задача — загрузить один из встроенных наборов данных Seaborn и построить несколько базовых графиков для анализа данных.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips') # или другой датасет на выбор

df.head()

print('Линейный график')
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x='day', y='total_bill', data=df, errorbar=None, estimator='mean', marker='o')
ax.set(
    title='Средняя сумма счета по дням',
    xlabel='День недели',
    ylabel='Средняя сумма счета',
)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print('Столбчатый график')
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x='day', y='tip', data=df, estimator=sum)
ax.set(
    title='Сумма чаевых по дням',
    xlabel='День недели',
    ylabel='Сумма чаевых',
)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print('Гистограмма')
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(df['total_bill'], bins=30, kde=True)
ax.set(
    title='Распределение суммы счета',
    xlabel='Сумма счета',
    ylabel='Частота',
)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print('Диаграмма рассеяния')
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='day', palette='viridis', s=100)
ax.set(
    title='Сумма счета vs Чаевые',
    xlabel='Сумма счета',
    ylabel='Чаевые',
)
plt.grid(True)
plt.tight_layout()
plt.show()
```
