### Задание 1: Основы программирования на Python
* Описание: Напишите функцию,которая принимает список чисел и возвращает список квадратов этих чисел

```python
def the_square_of_numbers(list_numbers: list) -> list:
    return [i*i for i in list_numbers]


the_square_of_numbers([1, 2, 3, 8, 9])
```

### Задание 2: Структуры данных в Python
* Описание: Напишите функцию, которая принимает строку и возвращает словарь, где ключами являются символы строки,а значениями–количество их вхождений в строку

```python
def processing_the_str(text: str) -> dict:
    return {char: text.count(char) for char in text.replace(' ', '')}


processing_the_str('hello, how are you?')
```

### Задание 3: Библиотеки
* Описание: Используйте встроенный датасет titanic из библиотеки seaborn и выполните следующие задания
1. Выведите первые 5 строк DataFrame.
2. Замените все пропущенные значения в столбце age средним значением поэтому столбцу.
3. Создайте новый столбец age_category, который будет содержать категориальные значения"Child", "Adult" и "Senior" на основе возраста(age).

```python
import seaborn as sns
import numpy as np
import pandas as pd
df = sns.load_dataset('titanic')

print('Первые 5 строк DataFrame')
df.head(5)
```

```python

print('Замена всех пропущенных значений в столбце age средним значением по этому столбцу')
df['age'].fillna(int(df['age'].mean()), inplace=True)
df['age'].isna().sum()
```

```python
print('Новый столбец age_category, который будет содержать категориальные значения"Child", "Adult" и "Senior" на основе возраста(age).')
age_category = [
    (df['age'] < 18),
    (df['age'] >= 18) & (df['age'] < 65),
    (df['age'] >= 65)
]
age_name = ['Child', 'Adult', 'Senior']

df['age_category'] = np.select(age_category, age_name, default='Unknown')

df.head(10)
```

### Задание 4: Манипуляции и анализ данных
* Описание: Продолжаем работу с датасетом titanic. Используя Pandas и датасет titanic из библиотеки seaborn, выполните следующие задания:
1. Выполните группировку данных по столбцу class и вычислите среднее значение для каждого класса по столбцу fare.
2. Постройте график распределения данных по столбцу age.
3. Постройте тепловую карту корреляционной матрицы числовых переменных

```python
print('Группировка данных по столбцу class и среднее значение для каждого класса по столбцу fare.')
df.groupby('class')['fare'].mean()
```

```python
import matplotlib.pyplot as plt

print('График распределения данных по столбцу age.')
fig, ax = plt.subplots(figsize=(14, 7))
age_values = df['age'].value_counts()
sns.barplot(x=age_values.index, y=age_values.values)
ax.set(
    title='График по возрасту',
    xlabel='Возраст',
    ylabel='Общее кол-во',
)
plt.tight_layout()
plt.xticks(ticks=range(0, len(age_values.index), 5))
plt.show()
```

```python
print('Тепловая карта корреляционной матрицы числовых переменных.')
plt.figure(figsize = (16,10))

sns.set(font_scale=1.4)

corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
corr_matrix = np.round(corr_matrix, 2)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Корреляционная матрица')
plt.show()
```
